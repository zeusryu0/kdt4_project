```python
import requests
import pandas as pd
import numpy as np
import html5lib
import json
from pandas.io.json import json_normalize
import time
```

```python
df = pd.DataFrame()
for page in range(1,1611):
    req=requests.get(f'http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json?key=8d0bb99dd87369ea48e84eb9ce3aa1f9&itemPerPage=100&curPage={page}')
    time.sleep(0.1)
    json_object = json.loads(req.text)
    df = pd.concat([df,json_normalize(json_object['peopleListResult']['peopleList'])])
```

```python
df.to_csv('./data/.csv')
```
