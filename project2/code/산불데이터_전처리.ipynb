```python
df.info()
```

```python
df['발생일시'] = df['발생일시_년']+'-'+ df['발생일시_월']+'-'+ df['발생일시_일']
df['발생일시']
```

```python
df['진화종료일시'] = df['진화종료시간_년']+'-'+ df['진화종료시간_월']+'-'+ df['진화종료시간_일']
df['진화종료일시']
```

```python
# 발생장소 발생장소_시도	발생장소_시군구	발생장소_읍면	발생장소_동리

df['발생장소'] = df['발생장소_시도'] + ' '+ df['발생장소_시군구']
df['발생장소']
```

```python
df.head()
```

```python
# 사용할 열 추출

df_fire = df[['발생일시','진화종료일시','발생원인','발생장소', '피해면적_합계']]
df_fire = df_fire.sort_values(by='피해면적_합계', ascending = False)
df_fire
```

### 피해면적에 따라 타겟설정

```python
# 날짜별 피해면적
df_firesum = df_fire.groupby('발생일시').sum()
df_firesum.sort_values(by='피해면적_합계',ascending=False,inplace=True)
df_firesum
```

```python
np.percentile(df_firesum, q = [25,50,75])
```

```python
df_firesum1 = df_firesum[df_firesum['피해면적_합계'] > 1.07 ]
df_firesum1.reset_index(inplace=True)
df_firesum1
```

### 지역별 타겟 개수

```python
df_fire['발생장소_광역시'] = df_fire['발생장소'].apply(lambda x: x.split(' ')[0])
df_fire
```

```python
df_fire['발생장소_광역시'].value_counts()
```

```python
# 가장 빈도수가 높은 지역

df_fire2= df_fire[df_fire["발생장소_광역시"]=='경북']
df_fire2
```

```python
# 중복값 제거
df_fire2 = df_fire2.drop_duplicates(subset = '발생일시')
df_fire2
```
