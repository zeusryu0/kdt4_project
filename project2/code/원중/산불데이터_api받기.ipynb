```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time

f_day = '19800101'
l_day = '20101231'

endday = []
endmonth = []
endtime = []
endyear = []

firecause = []

locbunji = []
locsi = []
locdong = []
locgungu = []
locmenu = []

startday = []
startdayofweek = []
startmonth = []
starttime= []
startyear = []

damagearea= []


page = 1
# totalCount>8477

for p in range(1, page+1):

    key = 'lj1bcK6kk1SlfbI0bHFwZNqo8sjEyQJ0XDwGOjV70SP2vXBS/DosbER90s3y1LOZCPEcQn3SqArgHpIv+Azj1Q=='

    url = 'http://apis.data.go.kr/1400000/forestStusService/getfirestatsservice'

    params = {'serviceKey' : key, 'numOfRows' : '10000', 'pageNo' : page , 'searchStDt' : f_day, 'searchEdDt' : l_day }
    
#     headers = {
#         'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/605.1.15 (KHTML,like GecKo) Chrome/79.0.3945.13 Safari/605.1.15',
#     }
    
    response = requests.get(url, params= params)
    time.sleep(0.3)
    res = response.text
    soup = BeautifulSoup(res,'html.parser')

    items = soup.find_all('item')
    for item in items:
        
        endday.append(item.find('endday'))
        endmonth.append(item.find('endmonth'))
        endtime.append(item.find('endtime'))
        endyear.append(item.find('endyear'))
        
        firecause.append(item.find('firecause'))
        
        locbunji.append(item.find('locbunji'))
        locdong.append(item.find('locdong'))
        locgungu.append(item.find('locgungu'))
        locmenu.append(item.find('locmenu'))
        locsi.append(item.find('locsi'))
        
        startday.append(item.find('startday'))
        startdayofweek.append(item.find('startdayofweek'))
        startmonth.append(item.find('startmonth'))
        starttime.append(item.find('starttime'))
        startyear.append(item.find('startyear'))
        
        damagearea.append(item.find('damagearea'))

        
df = pd.DataFrame({'startyear':startyear,'startmonth':startmonth,'startday':startday,'starttime':starttime,'startdayofweek':startdayofweek,
                   'endyear':endyear,'endmonth':endmonth,'endday':endday, 'endtime':endtime, 
                   'locbunji':locbunji,'locsi':locsi,'locgungu':locgungu,'locmenu':locmenu,'locdong': locdong,
                   'firecause':firecause, 'damagearea':damagearea
                 })

# df.to_csv('./data/산불_api.csv')      
```
