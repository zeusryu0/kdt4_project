```python
# 발생원인 합치기
lst = []

for i in range(len(df1)):
    if df1['발생원인_세부원인'][i] == '기타(직접입력)' :
        lst.append(df1['발생원인_기타'][i])
    else:
        lst.append(df1['발생원인_세부원인'][i])

df1.insert(17,'발생원인',lst)
df1
```
