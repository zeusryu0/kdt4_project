```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# ◆ 산불_피해면적기준

```python
df_sanbul = pd.read_csv('./data/산불타겟추가_피해면적기준.csv',index_col=0)
df_sanbul = df_sanbul.set_index('시간')
df_sanbul
```

```python
y = df_sanbul['산불발생여부']
x = df_sanbul.drop(columns = ['산불발생여부'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=111)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
from sklearn.metrics import classification_report, f1_score, roc_auc_score

svc = SVC(probability=True)
rfc = RandomForestClassifier()
ada = AdaBoostClassifier()
gbm = GradientBoostingClassifier()

xgb = XGBClassifier()
lgbm = LGBMClassifier()
cat = CatBoostClassifier()

# model1 = [svc, rfc, ada, gbm]
# model2 = [xgb, lgbm, cat]
model = [svc, rfc, ada, gbm,xgb, lgbm, cat]
```

```python
for m in model:
    m.fit(x_train,y_train)
    preds = m.predict(x_test)
    
    print(m.__class__.__name__)
    print(classification_report(y_test, preds))
    print(roc_auc_score(y_test,m.predict_proba(x_test)[:,1]))
    print('-'* 50)
```
