```python
# 재난일어난날을 바이너리로 만들어주기 
cold_high['재난일어난날'] = cold_high['재난일어난날'].fillna(0)
```

```python
# 고층기후 1990-01-01 부터 2021-12-31 까지
# 한파 1973-02-03 부터 2022-02-24 까지
# 한파와 고층기후의 null값이 0이기때문에 날짜를 맞추기위해 dropna를해줌 (이거 머지를 고층기후로 했기때문에 필요없음)
```

```python
# 재난이 일어난 시기를 2주 밀어서 예측시도
cold_high['재난일어난날'] = cold_high['재난일어난날'].shift(periods=14, axis=0)
```

```python
y = cold_high.재난일어난날
x = cold_high.drop('재난일어난날',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# 분류모델 import 
# Logistic Regression,SGD,KNN,SVM,Decision Tree, Random Forest, LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report,roc_auc_score,f1_score

lr = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC()
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier()
lgbm = LGBMClassifier()
```

```python
models = [lr,knn,dtc,rfc,lgbm]
# result = []
for i in models:
    i.fit(x_train,y_train)
    pred = i.predict(x_test)
    proba = i.predict_proba(x_test)[:,1]
    
    print(i.__class__.__name__)
    print('f1_score :', f1_score(y_test,pred))
    print('roc_auc_score :',roc_auc_score(y_test,proba))
    print('-'*50)
    
```

```python
#f1 score가 너무 낮게나왔으니 임밸런스 데이터를 제거하고 다시 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
import imblearn
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=0)
x_train_sm,y_train_sm = smote.fit_resample(x_train_scaled,y_train)
```

```python
y_train_sm.value_counts()
```

```python
models = [lr,knn,dtc,rfc,lgbm]
# result = []
for i in models:
    i.fit(x_train_sm,y_train_sm)
    pred = i.predict(x_test_scaled)
    proba = i.predict_proba(x_test_scaled)[:,1]
    
    print(i.__class__.__name__)
    print(classification_report(y_test,pred))
    print('roc_auc_score :',roc_auc_score(y_test,proba))
    print('-'*50)
    
```
