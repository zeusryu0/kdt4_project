```python
df['시간'] = df['시간'].astype('datetime64')
```

```python
# 옆으로 붙이기

#날짜 열 만들기
days = pd.date_range('1990-01-01','2021-12-31',freq='D')
df_c = pd.DataFrame({'시간':days})


areas = list(df['관측지점'].unique())
for area in areas:
    tmp = df[df['관측지점']== area].drop(columns='관측지점')
    tmp.columns = ['시간',f'{area}평균 기온', f'{area}최저 기온', f'{area}최고 기온', f'{area}1시간 최다강수량', f'{area}일강수량', f'{area}최대 순간풍속',
       f'{area}최대 순간 풍속 풍향', f'{area}최대 풍속', f'{area}최대 풍속 풍향', f'{area}평균 풍속', f'{area}풍정합', f'{area}최대 풍향', f'{area}평균 이슬점온도',
       f'{area}평균 상대습도', f'{area}평균 증기압', f'{area}평균 현지기압', f'{area}최고 해면 기압', f'{area}최저 해면기압', f'{area}평균 해면기압',
       f'{area}합계 일조 시간', f'{area}1시간 최다 일사량', f'{area}합계 일사', f'{area}일 최심신적설', f'{area}일 최심적설', f'{area}합계 3시간 신적설',
       f'{area}평균 전운량', f'{area}평균 중하층운량', f'{area}평균 지면온도', f'{area}합계 대형증발량', f'{area}합계 소형증발량', f'{area}안개 계속 시간']
    df_c = pd.merge(df_c,tmp, how='left', on='시간')
df_c 
```

```python
#사용할 도시

cities = ['신의주','평양','청진', '혜산', '강계', '함흥','블라디보스톡','도쿄', '삿포로', '센다이','가고시마','오사카', '후쿠오카','베이징', '선양', '창춘','상하이', '지난']


# 합치기(1996년 )
days = pd.date_range('1996-01-01','2021-12-31',freq='D')
df_c3 = pd.DataFrame({'시간':days})

for city in cities:
    tmp = df_w[df_w['도시명']== city].drop(columns=['도시명','군집'])
    tmp.columns = ['시간',f'{city}평균기온', f'{city}최고기온', f'{city}최저기온', f'{city}습도',f'{city}풍향', f'{city}풍속',
                   f'{city}강수량', f'{city}해면기압',f'{city}이슬점 온도']
    df_c3 = pd.merge(df_c3,tmp, how='left', on='시간')

df_c3
```

# 국내기후+ 고층기후 + 주변국기후(1996년~ 2021년 8월)

```python
df_c4 = pd.merge(df_c2, df_c3, how='left', on= '시간')
```
